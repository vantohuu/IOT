```python
df_sets = [k for k in os.listdir(DATASET_DIRECTORY) if k.endswith('.csv')]
```

```python
X_columns = [
    'flow_duration', 'Header_Length', 'Protocol Type', 'Duration',
       'Rate', 'Srate', 'Drate', 'fin_flag_number', 'syn_flag_number',
       'rst_flag_number', 'psh_flag_number', 'ack_flag_number',
       'ece_flag_number', 'cwr_flag_number', 'ack_count',
       'syn_count', 'fin_count', 'urg_count', 'rst_count',
    'HTTP', 'HTTPS', 'DNS', 'Telnet', 'SMTP', 'SSH', 'IRC', 'TCP',
       'UDP', 'DHCP', 'ARP', 'ICMP', 'IPv', 'LLC', 'Tot sum', 'Min',
       'Max', 'AVG', 'Std', 'Tot size', 'IAT', 'Number', 'Magnitue',
       'Radius', 'Covariance', 'Variance', 'Weight',
]
y_column = 'label'
```

```python
dataset = pd.DataFrame()
for df_set in df_sets:
    d = pd.read_csv(DATASET_DIRECTORY + df_set)
    sample_df = d.sample(n=500)
    dataset = pd.concat([dataset, sample_df])
```

```python
data = pd.DataFrame()
for df_set in df_sets:
    d = pd.read_csv(DATASET_DIRECTORY + df_set)
    data = pd.concat([data, d])
```

```python
att_list = data['label'].unique()
print(att_list)
```

```python
max = dataset['label'].value_counts().max()

for i in att_list:
    if (dataset['label'] == i).sum() < max:
        if (data['label'] == i).sum() >= max - (dataset['label'] == i).sum():
            samples = data[data['label'] == i].sample(n=max - (dataset['label'] == i).sum())
            dataset = pd.concat([dataset, samples])
        else:
            samples = data[data['label'] == i].sample(n=(data['label'] == i).sum() - (dataset['label'] == i).sum())
            dataset = pd.concat([dataset, samples])
```
